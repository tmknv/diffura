```python
#(2x+1)y'' + 4xy' -4y =0
```

```python
A=0
B=1
dx=0.01
```

```python
def __x__(i,x0,dx):
    return dx*i+x0
```

```python
def __y__(a0_x,a1_x,a2_x,dx,x_i,y_i,y_i_last):
    f_el=-a0_x*y_i*dx**2
    s_el=a1_x*y_i*dx
    th_el=(2*y_i-y_i_last)*a2_x
    return ((f_el+s_el+th_el)/(a2_x+a1_x*dx))
```

```python
def y_1(y,y_last,dx):
    return (y-y_last)/dx
```

```python
def y_2(y,y_last,y_last_last,dx):
    return (y_1(y,y_last,dx) -y_1(y_last,y_last_last,dx))/dx
```

```python
x0=0
lx=[]
ly=[]

lx.append(x0)

first_y=A+dx*B
ly.append(first_y)

print('y: ',ly,' x: ',lx)
```

```python
def calc(lx,ly,dx):
    i=1
    x=-float('inf')
    while x < B:

        x=__x__(i,x0,dx)

        a0_x=(2*x+1)
        a1_x=4*x
        a2_x=-4

        lx.append(x)

        y=__y__(a0_x,a1_x,a2_x,dx,lx[i-1],ly[i-1],ly[i-2])
        ly.append(y)

        i+=1
    print("First 3 element for x:", '\n', lx[:3],'\n',"Last 3 element for x:", '\n', lx[-3:], "\n" \
        "First 3 element for y:", '\n', ly[:3],'\n',"Last 3 element for y:", '\n', ly[-3:], "\n")
    return lx,ly

dx_i=10
while True:

    lx_old,ly_old=calc(lx,ly,dx/dx_i)
    dx_i*=10
    lx_new,ly_new=calc(lx,ly,dx/dx_i)
    if abs(ly_new[-1]-ly_old[-1])<10**-6:
        break

lx,ly=lx_new,ly_new
ly[-1]
```
